```python
import pandas as pd
import numpy as np
import plotly.express as pe
import plotly.express as px
import plotly.graph_objects as pg
import seaborn as sns
#sns.set(style = 'dargrid')
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
data = pd.read_csv("salaries_clean.csv")
data.head(3)
```

```python
#what data profession is the most paid in the US
col = data[['job_title','job_title_category', 'annual_base_pay','location_country']]
col = col.loc[(data['job_title_category'] == 'Data') & (data['location_country'] == 'US')]
col.groupby(['job_title_category','job_title','location_country'])[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = False)
```

```python
# return all data professionals that earns above 100k annually
data.loc[(data['job_title_category'] == 'Data') & (data['annual_base_pay'] > 100000)]
```

```python
# return all software professionals working in US and earning above 120k
data.loc[(data['job_title_category' ] == 'Software') & (data['location_country' ] == 'US') & (data['annual_base_pay'] >= 120000)]
```

```python
# return all the web developer in google that are living in the US
data[['job_title','job_title_category', 'employer_name','location_country']]
data.loc[(data['job_title'] == 'web developer') & (data['employer_name'] == 'google')  & (data['location_country'] == 'US')]
```

```python
# what is the average income all Management Workers in the US
data[['job_title_category','job_title','annual_base_pay','location_country']]
dt = data.loc[(data['job_title_category'] == 'Management') & (data['location_country'] == 'US')]
dt.groupby(['job_title_category','job_title','location_country'])[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending= True)
```

```python
#In Canada,what state has the highest paid software engineer
dy = data[['job_title','location_country','annual_base_pay','location_state']]
dt = dy.loc[(dy['location_country'] == 'CA') & (dy['job_title'] == 'software engineer') ]
dt.groupby(['job_title','location_country','location_state'])[['annual_base_pay']].mean().sort_values(by= 'annual_base_pay',ascending= False)
```

```python
#Choose 3 countries apart from US,show what country management workers pay the most
data[['job_title_category','job_title','location_country','annual_base_pay']]
dt= data.loc[((data['location_country'] == 'CA')|(data['location_country'] == 'FR')|(data['location_country'] == 'AU')) & (data['job_title_category'] == 'Management')]
dt.groupby(['job_title_category','job_title','location_country'])[['annual_base_pay']].mean().sort_values(by= 'annual_base_pay',ascending= False)
```

```python
# - What job would you advice me to venture into if I were relocating to US
data[['job_title_category','job_title','annual_base_pay','location_country']]
dt = data.loc[(data['location_country']) == 'US' ]
dt.groupby(['job_title_category','job_title','location_country'])[['annual_base_pay']].mean().sort_values(by= 'annual_base_pay', ascending=False)
dt.head(10)
```

```python
# - Who gets the most annual bonus between the Annual Bonus of Management Workers and Applied Science
df = data[['job_title_category','annual_bonus']]
df = df.loc[((data['job_title_category'] == 'Management')| (data['job_title_category'] == 'Applied Science'))  ]
df.groupby('job_title_category')[['annual_bonus']].mean().sort_values(by= 'annual_bonus', ascending=False)
```

```python
# - Granulate into Web Developers, Engineers, and Software Developers in the US. Show who is the least paid
data[['job_title','location_country','annual_base_pay']]
dt= data.loc[((data['job_title'] == 'Web developer')| (data['job_title'] == 'engineer') | (data['job_title'] == 'software developer')) & (data['location_country'] == 'US')]
dt.groupby(['job_title','location_country'])[['annual_base_pay']].mean().sort_values(by= 'annual_base_pay', ascending= True)
```

```python
# - Annual Bonus is an extra cost on Organisations. If you were an employer, show me the job title category you won't likely employ
dt = data[['job_title_category', 'annual_bonus']]
dt.groupby(['job_title_category'])[['annual_bonus']].mean().sort_values(by= 'annual_bonus', ascending= False)
```
